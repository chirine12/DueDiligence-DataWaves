```python
import os
import pdfplumber
import tqdm
from pathlib import Path
```

```python
# Fonction pour extraire le texte de chaque PDF dans le dossier
def extract_text_from_pdfs(pdf_dir):
    pdf_files = Path(pdf_dir).glob("*.pdf")  # Récupérer tous les fichiers PDF du dossier
    texts = []
    
    for pdf_file in pdf_files:
        with pdfplumber.open(pdf_file) as pdf:
            text = ""
            for page in pdf.pages:
                text += page.extract_text()  # Extraire le texte de chaque page
            texts.append((pdf_file.name, text))  # Ajouter le nom du fichier et le texte extrait

    return texts

# Exemple d'utilisation de la fonction
PDF_DIR = r"C:\Users\MSI\Downloads\PDFs"  # Remplace par ton propre chemin
texts = extract_text_from_pdfs(PDF_DIR)

# Afficher les textes extraits
for pdf_name, text in texts:
    print(f"Fichier: {pdf_name}")
    print(f"Texte extrait: {text[:300]}...")  # Afficher les 300 premiers caractères du texte extrait
    print("\n---\n")
```

```python
import requests
import json

# Fonction pour envoyer une requête à Ollama et obtenir des questions sur le contexte de la due diligence en cryptomonnaie
def ask_ollama_for_questions(context):
    ollama_url = "http://localhost:11434/v1/chat/completions"  # URL de l'API Ollama

    # Structure de la requête
    data = {
        "model": "mistral",  # Modèle Mistral
        "messages": [
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": f"Contexte : {context}\n\nPlease generate a list of relevant questions related to due diligence in cryptocurrency based on this context. Only provide the list of questions in English."}
        ]
    }

    headers = {"Content-Type": "application/json"}
    response = requests.post(ollama_url, headers=headers, data=json.dumps(data))

    if response.status_code == 200:
        return response.json()['choices'][0]['message']['content']
    else:
        return f"Erreur : {response.status_code}, {response.text}"

```

```python
def generate_questions_from_pdfs(PDF_DIR, output_md_file):
    texts = extract_text_from_pdfs(PDF_DIR)
    
    with open(output_md_file, "w", encoding="utf-8") as md_file:
        md_file.write("# Due diligence questions\n\n")

        # Pour chaque fichier PDF extrait, générer des questions basées sur son contenu
        for pdf_name, text in texts:
            print(f"Treating PDF File : {pdf_name}")
            questions = ask_ollama_for_questions(text)
            md_file.write(f"## File : {pdf_name}\n\n")
            md_file.write(f"{questions}\n\n")
            md_file.write("---\n\n")

    print(f"Questions saved in {output_md_file}")

```

```python
output_md_file = "questions_bank.md"

generate_questions_from_pdfs(PDF_DIR, output_md_file)
```
